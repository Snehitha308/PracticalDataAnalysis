```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, FunctionTransformer, OneHotEncoder
from sklearn.compose import ColumnTransformer
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv("stroke.csv")
```

```python
print(df.dtypes)
```

```python
df = df.dropna()
```

```python
y = df['stroke']
x = df.drop('stroke', axis=1)
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
symmetric_cols = ['hypertension', 'heart_disease']
skewed_cols = ['id', 'age', 'avg_glucose_level', 'bmi']
categorical_cols = ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, FunctionTransformer, OneHotEncoder
from sklearn.compose import ColumnTransformer
import numpy as np

symmetric_pipeline = Pipeline([
  ('imputer', SimpleImputer(strategy='mean')),
  ('scaler', StandardScaler())
])

skewed_pipeline = Pipeline([
  ('imputer', SimpleImputer(strategy='mean')),
  ('log_transform', FunctionTransformer(np.log1p)),
  ('scaler', StandardScaler())
])

categorical_pipeline = Pipeline([
  ('encoder', OneHotEncoder())
])

preprocessor = ColumnTransformer([
    ('numeric_symmetric', symmetric_pipeline, symmetric_cols),
    ('numeric_skewed', skewed_pipeline, skewed_cols),
    ('categorical', categorical_pipeline, categorical_cols)
])

pipeline = Pipeline([
    ('preprocessor', preprocessor)
])
```

```python
pipeline.fit(df)
train_transformed = pipeline.transform(train_x)
test_transformed = pipeline.transform(test_x)
```

```python
linear = LinearRegression()
linear.fit(train_transformed, train_y)

y_pred = linear.predict(test_transformed)

y_temp = []
for temp in y_pred:
  if temp > 0.5:
    y_temp.append(1)
  else:
    y_temp.append(0)

print(f"Accuracy: {accuracy_score(y_temp, test_y) * 100}%")
```

```python
neigh = KNeighborsClassifier(n_neighbors=3)
```

```python
neigh.fit(train_transformed, train_y)
```

```python
neigh_test = neigh.predict(test_transformed)
print(f"Accuracy of Logistic: {accuracy_score(neigh_test, test_y) * 100}%")
```
